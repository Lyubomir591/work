## Задание1
- 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки домаЕсли реновации дома не было, то в новом признаке поставьте 0
- 1.3 Создайте признаки года продажи, месяца продажи
- 1.4 Удалите признаки date, zipcode, lat, long

```python
df["delta_renovated"] = df["yr_renovated"]-df["yr_built"]
def ren(arg:int):
    
    if arg < 0 :
        return 0
    return arg
df['delta_renovated'] = df['delta_renovated'].apply(ren)
df.sample(5)
```

- 1.3 Создайте признаки года продажи, месяца продажи

```python
year = []
month = []
for i in df['date']:
    if i[:4].isdigit():
           year.append(i[:4])
    if i[4:6].isdigit():
        month.append(i[4:6])         
        
df['year of sale'] = year     
df['month of sale'] = month      
```

```python
price = df.pivot_table(index="bedrooms", values = "price" ,aggfunc="mean")
def toFixed(numObj, digits=0):
    return f"{numObj:.{digits}f}"
price["price"] = price['price'].apply(toFixed)
price["price"] = price["price"].astype("int64")
price.sort_values(["price","bedrooms"])
```
