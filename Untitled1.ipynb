 # **Задание1**

**На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе**

```python
sklad_frukt = ["Яблоки","Груши","Персики","Манго","Мандарины"]
frukt_cal = [15,22,44,15,4]

def stat_sum ( list1:list, list2:list ):
    fruct_dict = {}
    for i in range(len(sklad_frukt)):
            fruct_dict[list1[i]]=list2[i]
    return fruct_dict
book = stat_sum(sklad_frukt,frukt_cal)
def sum_value  (dict:dict):
    return sum(dict.values())
sum_book = sum_value(book)
print(f"Всего фруктов на складе: {sum_book} шт.")
```

# **Задание2**

**Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200]
Используйте list comprehensions**

```python

def sort (arg):
    price = [i for i in arg if i>=0]
    return price
salary = [100, 125, -90, 345, 655, -1, 0, 200]
price = sort(salary)
print( f"Сумарные затраты : {sum(price)}")
    
```
